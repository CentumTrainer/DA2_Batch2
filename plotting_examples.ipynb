```python
#setting label of x and y axis & also title chart
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50]
y = [65, 98, 170, 220, 310]
plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Over wise Runs scored \n India Vs Australia')
plt.plot(x,y)
plt.show()
```

```python
#changing line color and line width & style
#setting label of x and y axis & also title chart
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50]
y = [65, 98, 170, 220, 310]
plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Over wise Runs scored \n India Vs Australia')
plt.plot(x,y,'r',linewidth=5, linestyle='dashed')
plt.show()
```

```python
#Changing marker type,size & color
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50]
y = [65, 98, 170, 220, 310]
plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Over wise Runs scored \n India Vs Australia')
plt.plot(x,y,marker='+', markersize='15', markeredgecolor='red')
plt.show()
```

# Bar Graph Ploting

```python
#Bar graph used to represnt data in the form of vertical or horizontal bars.
#usefult to compare quantities
import matplotlib.pyplot as plt

OverRange=['1-10','11-20','21-30','31-40','41-50']
RunsScored=[65,55,70,50,80]

plt.bar(OverRange, RunsScored)

plt.xlabel('Over Intervals')
plt.ylabel('Runs Scored')
plt.show()

```

```python
#change bar widht & color
import matplotlib.pyplot as plt

OverRange=['1-10','11-20','21-30','31-40','41-50']
RunsScored=[65,55,70,50,80]

plt.bar(OverRange, RunsScored,width=0.3,color=['r','g','b','c','m'])

plt.xlabel('Over Intervals')
plt.ylabel('Runs Scored')
plt.show()
```

```python
#Bar graph horizontal bars.

import matplotlib.pyplot as plt

OverRange=['1-10','11-20','21-30','31-40','41-50']
RunsScored=[65,55,70,50,80]

plt.barh(OverRange, RunsScored)

plt.xlabel('Over Intervals')
plt.ylabel('Runs Scored')
plt.show()
```

# Multiple Bars

```python
import matplotlib.pyplot as plt
import numpy as np

teamA = [50,60,70,80,90]
teamB = [55,65,60,75,96]

x = np.linspace(1,51,5)
plt.bar(x,teamA,width=3,color='g',label='Australia')
plt.bar(x+3,teamB,width=3,color='yellow',label='India')
plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Scoring chart Australia Vs India')
plt.legend()
plt.show()
```
